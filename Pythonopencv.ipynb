```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image1=cv2.imread("/content/cloud.webp")
plt.imshow(image1[:,:,::-1]);plt.title("Original image:");plt.axis("off")

image1_copy=image1.copy()
gray=cv2.cvtColor(image1_copy,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray");plt.title("Gray image:");plt.axis("off")
grayinv=cv2.bitwise_not(gray)
plt.imshow(grayinv,cmap="gray");plt.title("Bitwise Inverted:");plt.axis("off")
_,binary=cv2.threshold(grayinv,100,255,cv2.THRESH_BINARY)
plt.imshow(binary,cmap="gray");plt.title("Binary Image:");plt.axis("off")
contours,hierarchy=cv2.findContours(binary,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
image1_copy=cv2.drawContours(image1_copy,contours,-1,(0,0,255),2)
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image1_copy[:,:,::-1]);plt.title("Contour Image with Thresholding:");plt.axis("off")
```
